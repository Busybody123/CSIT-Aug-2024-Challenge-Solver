Given that the s-box has been modified to be linear, AES(key).encrypt(msg) is a linear function for the key and message.


Therefore, AES(key).encrypt(msg) = Linear(key, msg) $\oplus$ tweak.
- $E(p,k,t) = F(p) \oplus G(k) \oplus H(t)$

We also have a decryption function $D_{k,t}(E(p,k,t)) = p$.


Our goal is to obtain $E(flag,0,t)$.

Since this is a known-plaintext attack, we can choose
- $pad = E(pad,0,t) = F(pad) \oplus G(0) \oplus H(t)$

Coincidentally, ciphertext[6] is 64 bytes, so the last block is padding.
- $encPad = E(pad,k,t) = F(pad) \oplus G(k) \oplus H(t)$

We also have the encrypted flag, 
- $encFlag = E(flag,k,t) = F(flag) \oplus G(k) \oplus H(t)$


We then use XOR's self-inverse property to cancel them out.

$encPad \oplus encFlag = F(pad) \oplus G(k) \oplus H(t) \oplus F(flag) \oplus G(k) \oplus H(t)$
                      $= F(pad) \oplus F(flag)$

$pad \oplus encPad \oplus encFlag = F(pad) \oplus G(0) \oplus H(t) \oplus F(pad) \oplus F(flag)$
                                 $= F(flag) \oplus G(0) \oplus H(t)$


We then use the decryption function to decrypt the flag.

$D_{0,t}(E(flag,0,t)) = flag$
